Transforming from Cartesian Coordinates to polar coordinates

$\mathbf{e_r} = \cos(\psi)\mathbf{e_x} + \sin(\psi)\mathbf{e_y}$

and 

$\mathbf{e_\psi} = -\sin(\psi)\mathbf{e_x} + \cos(\psi)\mathbf{e_y}$

$\mathbf{e_\psi}$ is referred to as the tangential direction. rotates in the positive-z direction

# 

```python
import numpy as np
import matplotlib.pyplot as plt
theta= np.linspace(0,2*np.pi*10,1000)
R = 10
h=2
psi=0.0

x = R*np.cos(-theta+psi)
y = R*np.sin(-theta+psi)
z = h/(2*np.pi)*theta
ax = plt.axes(projection='3d')
ax.plot3D(x,y,z)

```

$\Gamma_{tot}$ is the total circulation of the rotor. $\Gamma_b=\Gamma_h=\Gamma_tot/n_B$, where $n_B$ is the total number of blades. The implication is that each blade contributes one third of the total circulation of the rotor. Note further that $\Gamma_b$ and $\Gamma_h$ are displayed in Figure 1 of Branlard(2015). They are, respectively, a blade component and a helical component. the blade itself seems to be modeled as a vortex line, and then the helix is a vortex line. Coleman is noted to have decomposed these circulations into a circular component and linear vortex filaments. $\Gamma_r$  is a single vortex filament with $\Gamma_r=-\Gamma_{tot}\mathbf{e_z}$. $\Gamma_l=\Gamma_b\mathbf{e_z}$ and $\Gamma_t = -\Gamma_b\mathbf{e_\psi}$

This seems to imply that the total circulation is zero, with a "root" circulation that is equal and opposite to the total circulation from the blades.
